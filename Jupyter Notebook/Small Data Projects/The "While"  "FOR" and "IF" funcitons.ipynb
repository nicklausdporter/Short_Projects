```python
import numpy as np
from numpy.random import randn
```

```python
randn()
```

```python
#---- -2 ---- -1 ---- 0 ---- 1 ---- 2 ----
answer = None
x = randn()
if x > 1:
    answer = "Greater than 1"
print (x)
print (answer)
```

```python
#---- -2 ---- -1 ---- 0 ---- 1 ---- 2 ----
answer = None
x = randn()
if x > 1:
    answer = "Greater than 1"
else:
    answer = "Less than 1"
print (x)
print (answer)
```

```python
#---- -2 ---- -1 ---- 0 ---- 1 ---- 2 ----
#nested statements
answer = None
x = randn()
if x > 1:
    answer = "Greater than 1"
elif x >= -1:
    answer = "Between -1 and 1"
else:
    answer = "Less than -1"
print (x)
print (answer)
```
